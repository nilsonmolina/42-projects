```python
# 3. Write a function that creates an incremental N-dimensional (nd) array of 
# dimension (1,n) with values between 0 and 1. Use arr.shape to verify
def create_array(n=2):
    return np.linspace(0, 1, n).reshape(1, n)
ARR = create_array(13)
print(ARR)
print("arr shape: ", ARR.shape)
print("arr type:  ", type(ARR))
```

```python
# 5. Using the function in Q3, get m vectors and bind them together (to have a m x n) 
# matrix. Plot the matrix with matplotlibs imshow"
def create_matrix(m, n):
    matrix = [create_array(n) for i in range(m)]
    return np.vstack(matrix)

ARR = create_matrix(3, 3)
print(ARR)
plt.imshow(ARR)
```

```python
# 6. Multiply the resulting matrix from Q5 with the matrix of a picture of your 
# choice. Plot the resulting matrix.
IMG = plt.imread('img.png')
rows, columns, channels = IMG.shape[0], IMG.shape[1], IMG.shape[2]

NEW = np.copy(IMG)
FILTER = create_matrix(rows, columns)
for i in range(channels):
    NEW[:, :, i] *= FILTER
    
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
axs[0].axis('off')
axs[0].imshow(IMG)
axs[1].axis('off')
axs[1].imshow(NEW)
```
